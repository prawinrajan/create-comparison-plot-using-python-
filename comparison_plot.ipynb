```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


methods = ["Logistic Regression", "KNN", "SVM", "Naive Bayes", "Decision Tree", "Random Forest","neural-network"]
accuracy = [86.88,86.89,87.13,83.06,100,85.25,99.56]
colors = ["purple", "green", "orange", "magenta","#CFC60E","#0FBBAE","red"]

sns.set_style("whitegrid")
plt.figure(figsize=(20,5))
plt.yticks(np.arange(0,100,10))
plt.ylabel("Accuracy %")
plt.xlabel("Algorithms")
sns.barplot(x=methods, y=accuracy, palette=colors)
plt.savefig('using_normmalization.png')
plt.show()
```
